```python
from birdclef.plot_motif import plot_melspectrogram, plot_cens
import matplotlib.pyplot as plt
from pathlib import Path


train_audio = Path("../data/raw/birdclef-2022/train_audio")


def plot(examples):
    for path in examples:
        fig, ax = plt.subplots(
            nrows=2, ncols=2, sharex=False, sharey=False, figsize=(7.5, 5)
        )
        name = path.name.split(".")[0]
        species = path.parent.name
        fig.suptitle(f"Spectrogram and matrix profile for species {species} ({name})")
        plot_melspectrogram([ax[0, 0], ax[1, 0]], path, mp_window=80 * 5)
        plot_cens([ax[0, 1], ax[1, 1]], path, cens_sr=10, mp_window=50)
        plt.show()


bird = "puaioh"
tracks = list(train_audio.glob(f"{bird}/*.ogg"))
print(tracks)
plot(tracks[:2])
```

```python
import random

bird = "houfin"
random.seed(2022)
tracks = random.choices(list(train_audio.glob(f"{bird}/*.ogg")), k=3)
print(tracks)
plot(tracks)
```
