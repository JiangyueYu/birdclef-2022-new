```python
from pathlib import Path

path = Path("../data/intermediate/2022-03-12-motif-triplets-5e+05")
items = list(path.glob("*.npy"))
len(items)
```

```python
import random
import IPython.display as ipd
import matplotlib.pyplot as plt
import numpy as np

for i in range(5):
    x = random.choice(items)
    y = np.load(x)
    plt.plot(y)
    plt.show()
    ipd.display(ipd.Audio(y, rate=32000))
```
