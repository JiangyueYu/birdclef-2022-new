```python
from pathlib import Path
import json
import pandas as pd

scored_birds_path = "../data/raw/birdclef-2022/scored_birds.json"
scored_birds = set(json.loads(Path(scored_birds_path).read_text()))
df = pd.read_parquet("../data/processed/2022-04-03-motif-consolidated.parquet")
df["species"] = df.source_name.apply(lambda x: x.split("/")[1])
aggregated = (
    df[["species", "duration_seconds"]][df.species.isin(scored_birds)]
    .groupby("species")
    .agg(["count", "sum", "median", "mean", "std"])
)
aggregated
```
